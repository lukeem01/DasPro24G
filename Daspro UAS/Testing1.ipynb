```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("rounded_hours_student_scores.csv")
df.head(10)
```

```python
df.describe()
```

```python
# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Hours'], df['Scores'], color='green', edgecolors='black')
plt.title("Scatter Plot: Hours vs Scores (Matplotlib)")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)
plt.show()

# Heatmap Korelasi
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap (Seaborn)")
plt.show()
```

```python
X = df[['Hours']]
y = df['Scores']

model = LinearRegression()
model.fit(X, y)

print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_[0]}')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Hours'], y=df['Scores'], color='green', label='Data Asli')
plt.plot(df['Hours'], model.predict(X), color='red', label='Garis Regresi')
plt.title("Regresi Linier: Hours vs Scores (Seaborn + Matplotlib)")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")
```
